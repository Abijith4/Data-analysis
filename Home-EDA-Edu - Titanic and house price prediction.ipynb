```python
#detecting outliers using Z-Score----z=(x-mean)/std
```

```python
outliers=[]   #ouliers as datas or list
def detect_outliers(data): # creating the function which detscts the oulier and  to detect the outliers with the help of z-score we have to find the z-score value    
    
    threshold=3 #if the data is falling in the standard deviation if the data is falling i am not going to consider it has ouliers, if it is falling away from the 3 standard deviation we are going to consider it as oulier  
    mean=np.meanean(data) # 1st i am going to compute the mean 
    std=np.stdtd(data)# then the standard deviation 
    
    # then i am going to put a forloop in to data set 
    for i in data:
        z_score=(i-mean)/std # for each and every dataset i am going to subtaract X-mean to find the z-score , whatever the z-score value commng over here. 
        if np.abs(z_score)>threshold: # if the z-score value is greater than the threshold value,for eg: if its falling away form the 3 standard deviation. i am going to append that i value  in that outliers 
            outliers.append(i) #if its falling away form the 3 standard deviation. i am going to append that i value  in that outliers
            # This particular i values depends that it does not falls in the 3rd std deviation so i am adding that in the outliers list it will check each and every values into the dataset  
    return outliers
```

```python
# What is regression ?
#      Regression analusis is a form of predictive modelling technique which investigates the relationship between a dependent and independent variable.
#     ii. A regeression analysis involves graphying a line over a set of data points that most closely fits the overall shape of the data or regression shows the changes in a dependent variable in the y- axis to the changes in the explanatory variable on the x-axis.
# What are the uses of regression?
#      Three major uses for regression analysis are 
#          i. Determining the strength of predictors- The first being determining the strength of predictors, the regression might be used to identify the strength of the effect that the independent variables have on the dependent variable.  
#          ii.Forecasting an effect and - In this the regression can be used to forecast effects or impact of changes. That is the regression analysis helped us to understand how much the dependent variable changes with the change in one or more independent variable   
#          iii.Trend forecasting- In this regression analysis to predict trends and future values, the regression analysis can be used to get point estimates 
```

```python
#Linear regression Selection criteria
#     i. classification and Regression capabilities 
#         Regression models predict continuous variable such as sales made on a day or preddict the temperature of acity their Reliance on a polynomial like a straight line to fit a data set poses a real challenge when it comes towards building a classification capability    
#     ii. Data quality 
#         Remove one data point that could optimize the regression in simple linear regresion the outliers can significantly disrupt the outcomes just for now,you can know that if you remove the outliers your model will become very good
#     iii. Computational complexity 
#           A linear regression is often not computationally expensive as compares to the decision tree or the clustering algorithm the order of complexity for n training example and x features usually falls in either Big O of X square or bigger of xn next is comprehensible 
#      iV. Comprehensible and Transparent 
#            The linear regression are easily comprehensible and transparent in nature.They can be represented by a simple mathematical notation to anyone and can be understood very easily so these are some of the criteria based  which we are selecting the linear regression algorithm 
```

```python
# Mean X and Y 
mean_x = np.mean(x)
mean_y = np.mean(y)

# Total number of values 
m=len(x)

#using the formula to calculate b1 and b2 

numer=0
denom=0 

for i in range(m):
    numer+=(x[i]-mean_x)*(y[i]-mean_y)
    denom+=(x[i]-mean_x)**2
b1=numer/denom
b0=mean_y-(b1*mean_x)

#print coefficients 
print(b1,b0)
```

```python
# plotting values and regression line 

max_x = np.max(x) 
min_x = np.min(x) 

#Calculating line values x and y 

x= np.linspace(min_x,max_x)
y=b0 + b1 * x
# Plotting line 
plt.plot(x,y,color='#58b970', label='Regression line')

#ploting scatter points 

plt.scatter(x,y,c='#ef5423',label='scatter plot')

plt.xlabel('Price according to bedroom')
plt.ylabel('Housing Price')

plt.legend()
plt.show()
```

```python
total_features=[]
total_features_name=[]
selected_correlation_value=[]
r2_score=[]
mae_value=[]
mse_value=[]
```

```python
# python method for the performance matrix,Why because we will be using this many times 
```

```python
def performance_metrics(features, th, y_true, y_pred):
    score = r2_score(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    total_features.append(len(features)-1)
    total_features_name.append(str(features))
    selected_correlation_value.append(th)
    r2_scores.append(score)
    mae_value.append(mae)
    mse_value.append(mse)
    
    metrics_dataframe = pd.DataFrame(data= [total_features_name, total_features, selected_correlation_value, r2_scores, mae_value, mse_value],
                                    index = ['features name', '#feature', 'corr_value', 'r2_score', 'MAE', 'MSE'])
    return metrics_dataframe.T
```

```python
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None, n_jobs=None, train_sizes=np.linspace(.1, 1.0, 10)):
    
    plt.figure()
    plt.title(title)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt

X = correlated_data.drop(labels = ['Price'], axis = 1)
y = correlated_data['Price']

title = "Learning Curves (Linear Regression) " + str(X.columns.values)

cv = ShuffleSplit(n_splits=100, test_size=0.2, random_state=0)

estimator = LinearRegression()
plot_learning_curve(estimator, title, X, y, ylim=(0.7, 1.01), cv=cv, n_jobs=-1)
plt.show()
```
